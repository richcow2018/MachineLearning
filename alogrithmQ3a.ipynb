```python
#Starting implementation
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Reading dataset and setting columns
df = pd.read_csv("/Users/jimmychu/Downloads/Lab_2/data/cardiac.csv")
df
```

```python
x = df.values[:, 0:31]
x = df.drop(['death'], axis=1)
y = df.values[:, 24]
#x=sum(x1,x2)
#x = x1 + x2
#x.append(x1,x2)
#x.extend(x2)
#x = df.values[:, 0:24]
feature_names = ['bhr', 'basebp', 'basedp', 'pkhr', 'sbp', 'dp', 'dose', 'maxhr','%mphr(b)', 'mpb', 'dpmaxdo', 'dobdose', 'age', 'gender', 'baseEF', 'dobEF', 'chestpain', 'posECG', 'equivecg', 'restwma', 'posSE', 'newMI', 'newPTCA','newCABG', 'hxofHT', 'hxofdm', 'hxofcig', 'hxofMI', 'hxofPTCA', 'hxofCABG']
trainX, testX, trainY, testY = train_test_split( x, y, test_size = 0.3, random_state=0)
```

```python
#Step 1 - Choosing Model - Import class which is planned to use
from sklearn import tree
```

```python
#Step 2 - Instantiate the estimator
dtree = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
#Step 3 - Fit model with data (Learning)
dtree.fit(trainX, trainY)
class_names = np.unique([str(i) for i in trainY])
```

```python
#Step 4 - Predict the response for the testing set
Y_pred = dtree.predict(testX)
print(Y_pred)
print("Accuracy: \n", dtree.score(testX, testY))
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, filled=True, rounded=True, special_characters=True, class_names=class_names, feature_names=feature_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
